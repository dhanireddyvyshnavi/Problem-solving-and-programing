### Handling Missing Values using Scikit-Learn

```python
#from  sklearn import impute.simpleimputer
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values="NaN",strategy="mean",axis=0)
```

```python
df
```

```python
df.iloc[:]
```

```python
df.iloc[:1]
```

```python
df.iloc[:,1:3]
```

```python
imputer=imputer.fit(df.iloc[:,1:3])
```

```python
df.iloc[:,1:3]=imputer.transform(df.iloc[:,1:3])
```

```python
df
```

```python
type(df.tempareture[0])
```

#### As imputer object works only on the numerical values,for the categorical data we deal with mode

#### Mode Imputation

```python
df.event.mode()
```

```python
df.event.value_counts()
```

```python
col=['event']
df[col]=df[col].fillna(df.mode().iloc[0])
df[col]
```

```python
df.fillna(df.mode())
```

#### Median Imputation

```python
df.fillna(df.median())
```
